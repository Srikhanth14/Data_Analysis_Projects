## **Importing The Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Data Collection and Pre-Processing**

```python
#loading the dataset
data=pd.read_csv('adult.csv')
```

```python
#Display top 10 rows of the dataset
data.head(10)
```

```python
#Display last 5 rows of the dataset
data.tail()
```

```python
#Information about the dataset
data.info()
```

```python
#Shape of the dataset
print("No. of rows in the dataset =",data.shape[0])
print("No. of columns in the dataset =",data.shape[1])
```

```python
#Checking for null values in the dataset
print("Are there any null values available in the dataset =",data.isnull().values.any())
```

```python
#checking for duplicated values
print('Is there any presence of duplicated values =',data.duplicated().values.any())
```

```python
#Dropping the duplicated values
data.drop_duplicates(inplace=True)
```

```python
#Checking for the presence of unknown values in the dataset
data.isin(['?']).sum()
```

```python
#Replace '?' with np.nan
data=data.replace('?',np.nan)
```

```python
data.isin(['?']).values.sum()
```

```python
#Again Checking for null values in the dataset
print("Are there any null values available in the dataset =",data.isnull().values.any())
```

```python
#Dropping the null values
data.dropna(how='any',inplace=True)
```

```python
#Again checking for duplicated values
print('Is there any presence of duplicated values =',data.duplicated().values.any())
```

```python
#Dropping the irrelevant features
data.drop(columns=['fnlwgt','educational-num'],axis=1,inplace=True)
```

```python
#Information about the dataset
data.info()
```

```python
#Summary of the Dataset
data.describe()
```

## **Exploratory Data Analysis**

```python
data.columns
```

```python
data.age.unique()
```

```python
#Distribution of income across different age groups
plt.figure(figsize=(12,10))
sns.histplot(x='age',hue='income',data=data,kde=True,palette=["#003f5c","#00ffff"])
plt.title('Income Distribution Across Age Group')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,8))
# Correlation between age and income
plt.subplot(121)
sns.boxplot(x='income', y='age', data=data,palette="Paired")
plt.title("Boxplot: Income vs Age")
plt.xlabel("Income")
plt.ylabel("Age")

# Correlation between age and income
plt.subplot(122)
sns.barplot(x='income', y='age', data=data,errorbar=None,palette="Paired")
plt.title("Barplot: Income vs Age")
plt.xlabel("Income")
plt.ylabel("Age")
plt.show()
```

```python
#Distribution of Income
data.income.value_counts()
```

```python
 plt.figure(figsize=(10,8))
 sns.countplot(x='income',data=data,palette=["#e27c7c", "#a86464"])
 plt.title("Distribution of Income")
 plt.xlabel('Income')
 plt.ylabel('Count')
 plt.show()
```

```python
#Covert income categorical to numeric assign 0 to <50K and 1 to >=50K
data['encoded_income'] = data['income'].map({'<=50K':0, '>50K':1}).astype('int')
```

```python
data['encoded_income'].value_counts()
```

```python
#Distribution of income across different workclass
work_income=data.groupby('workclass')['income'].value_counts()
work_income
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='workclass',hue='income',data=data,palette=["#599e94", "#6cd4c5"])
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Distribution of income across different Workclass')
plt.show()
```

```python
#Distribution of Income with Respect to Race
race_income=data.groupby('race')['income']. value_counts(normalize=True).unstack()*100
race_income=race_income.sort_values(by='>50K',ascending=False)
race_income
```

```python
fig,ax = plt.subplots(figsize=(10,8))
race_income.plot(kind='barh',color=["#599e94", "#6cd4c5"],ax=ax)
ax.set_title('Income Distribution Across Race')
ax.set_xlabel('Workclass')
ax.set_ylabel('Income')
plt.show()
```

```python
#Distribution of Income Across Relationship
data.relationship.value_counts()
```

```python
plt.figure(figsize=(12,10))
sns.countplot(x='relationship',hue='income',data=data,palette=["#115f9a", "#00ffff"])
plt.title("Distribution of Income Across Relationship")
plt.xlabel("Relationship")
plt.ylabel("Income")
plt.show()
```

## **Hypothesis**

## Hypothesis-1

## Individuals with higher education levels and who are married tend to have higher income levels.

```python
#Distribution of Income across Education
edu_income=data.groupby('education')['income'].value_counts(normalize=True).unstack()*100
edu_income=edu_income.sort_values(by='>50K',ascending=False)
edu_income
```

```python
fig,ax=plt.subplots(figsize=(12,10))
edu_income.plot(kind='barh',color=["#a4a2a8", "#df8879"],ax=ax)
ax.set_title('Income Distribution Across Educational Level')
ax.set_ylabel('Education')
ax.set_xlabel('Income Percentage ')
plt.show()
```

```python
data.columns
```

```python
#Distribution of Income based on Marital Status
mar_income=data.groupby('marital-status')['income'].value_counts(normalize=True).unstack()*100
mar_income=mar_income.sort_values(by='>50K',ascending=False)
mar_income
```

```python
fig,ax=plt.subplots(figsize=(10,8))
mar_income.plot(kind='barh',color=["#d7658b", "#c80064"],ax=ax)
ax.set_title("Income Distribution Across Marital Status")
ax.set_xlabel("Income Percentage")
ax.set_ylabel("Marital Status")
plt.show()
```

```python
#End of first hypothesis
plt.figure(figsize=(20,10))
sns.barplot(x='education',y='encoded_income',hue='marital-status',data=data,errorbar=None)
plt.title('Income Distribution across Educational level and Marital Status')
plt.xlabel('education')
plt.ylabel('income')
plt.show()
```


##Conclusion
##The First Hypothesis is supported as Individuals with higher education levels and who are married tend to have higher income levels.

##Hypothesis-4
##There is a positive correlation between the number of hours worked per week and income levels.


```python
data['hours-per-week'].unique()
```

```python
data['hours-per-week'].value_counts()
```

```python
#Distribution of Income based on Worked Hrs
plt.figure(figsize=(10,8))
sns.barplot(x='income',y='hours-per-week',data=data,errorbar=None,palette=["#fd7f6f","#8bd3c7"])
plt.title('Income Distribution based on Hours Worked')
plt.xlabel('Income')
plt.ylabel('Hours-Worked-per-Week')
plt.show()
```

```python
hrs_corr=data[['hours-per-week','encoded_income']]
hrs_corr=hrs_corr.corr(numeric_only=True)
hrs_corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(hrs_corr,annot=True,cmap='plasma')
plt.title('Hours Worked per Week Vs. Income')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='income', y='hours-per-week',data=data,palette=["#e2e2e2", "#e1a692"])
plt.title("Hours Worked per Week Vs. Income")
plt.xlabel("Income")
plt.ylabel("hours Worked per week")
plt.show()
```

##Conclusion
##The Fourth Hypothesis is supported as There is a positive correlation between the number of hours worked per week and income levels.


##Hypothesis-5

##Individuals with significant capital gains are more likely to have higher income levels, while those with substantial capital losses may experience lower income.


```python
#Correlation Matrix for Capital gain,loss and income
capital_gain_loss = data[['capital-gain', 'capital-loss', 'income','encoded_income']]
capital_gain_loss=capital_gain_loss.corr(numeric_only=True)
capital_gain_loss
```

```python
#Heatmap for the Correlation Matrix
plt.figure(figsize=(10,8))
sns.heatmap(capital_gain_loss,annot=True,cmap='jet')
plt.title('Capital Gain/Loss vs. Income')
plt.show()
```

```python
# Bar plots for average capital gain and capital loss with respect to income
plt.figure(figsize=(12, 10))
plt.subplot(121)
sns.barplot(x='income', y='capital-gain', data=data, palette=["#599e94", "#6cd4c5"],errorbar=None)
plt.title("Average Capital Gain by Income")
plt.xlabel('Income')
plt.ylabel('Capital-gain')

plt.subplot(122)
sns.barplot(x='income', y='capital-loss', data=data, errorbar=None,palette=["#a86464","#e27c7c"])
plt.title("Average Capital Loss by Income")
plt.xlabel('Income')
plt.ylabel('Capital-loss')
plt.show()
```

##Conclusion
##The Fifth Hypothesis is partially supported the individuals with significant capital gains are more likely to have higher income levels, while the correlation with substantial capital losses and lower income is not as strong.


##Hypothesis-6
##There are variations in income levels based on the native country or region, with certain regions having higher average incomes



```python
data.columns
```

```python
native_income=data.groupby('native-country')['income'].value_counts(normalize=True).unstack()*100
native_income=native_income.sort_values(by='>50K',ascending=False)
native_income
```

```python
fig,ax=plt.subplots(figsize=(15,8))
native_income.plot(kind='bar',color=["#00bfff","#115f9a"],ax=ax)
ax.set_title("Income Distribution based on Native Country")
ax.set_xlabel('Native-Country')
ax.set_ylabel('Income')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
 # Bar plot for income distribution across native countries
plt.figure(figsize=(12, 10))
sns.barplot(y='native-country', x='encoded_income', data=data,errorbar=None,order=data['native-country'].value_counts().index)
plt.title("Income Distribution Across Native Countries")
plt.xlabel("Native Country")
plt.ylabel("Income")
plt.show()
```

##Conclusion
##The Sixth Hypothesis is supported as there are variations in income levels based on the native country or region, with certain regions having higher average incomes.
